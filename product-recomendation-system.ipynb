**product recomendation based on similar products**

```python
df2=pd.read_excel("ds2.xlsx")
```

```python
df2.head
df2.isnull()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
df2['CustomerID'].isna().sum()
df2_mod = df2.dropna(subset=['CustomerID'])
df2_mod.head
```

```python
cim = df2_mod.pivot_table(
    index='CustomerID', 
    columns='StockCode', 
    values='Quantity',
    aggfunc='sum'
)
cim =cim.applymap(lambda x: 1 if x > 0 else 0)
```

```python
ism = pd.DataFrame(
    cosine_similarity(cim.T)
)
```

```python
ism.columns = cim.T.index
ism['StockCode'] = cim.T.index
ism = ism.set_index('StockCode')
sitms = list(ism.loc[22722].sort_values(ascending=False).iloc[:4] .index)
sitms
```

```python
df2_mod.loc[df2_mod['StockCode'].isin(sitms), ['StockCode', 'Description']].drop_duplicates().set_index('StockCode').loc[sitms]
```
